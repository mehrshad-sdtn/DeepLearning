```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))

def logistic_loss(y, y_hat):
    return -(y * np.log(y_hat) + (1 - y) * np.log(1 - y_hat))
```

```python
class Neuron():
  def __init__(self, ins, activation):
    self.inputs_c = ins
    self.activation = activation
    self.weights = np.random.randn(self.inputs_c, 1)
    self.bias = 0

  def __repr__(self):
    return f"Node({self.inputs_c}, {self.activation})"

  def forward(self, x):
    z = np.dot(x, self.weights) + self.bias
    y_pred = self.activation(z)
    return y_pred
```
